```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('100_drivers_data.csv')
```

```python
target_road_boundaries = np.array([[108.94765628015638, 34.23940650580562],
                                 [108.94615212765522, 34.23940384395769],
                                 [108.94615156073496, 34.2324012260476],
                                 [108.94765571288106, 34.232403887889106],
                                  ])
```

```python
df_road = pd.DataFrame(target_road_boundaries,columns = ['Longitude','Latitude'])
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x=df['Longitude'],y=df['Latitude'],s=1)
plt.plot(df_road['Longitude'],df_road['Latitude'],'-o',color='r')
plt.show()
```
