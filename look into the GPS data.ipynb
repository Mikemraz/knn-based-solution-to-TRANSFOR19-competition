```python
def get_df_of_some_drivers(df,num=10):
    """get the dataframe of some number of drivers sampled from original dataframe.
    
    Parameters:
    df: the dataframe to be sampled from.
    num: the number of drivers to sample
    
    Returns:
    df_of_some_drivers: dataframe of drivers of given number.
    
    """
    driver_ids = list(df['Driver_ID'].unique())
    selected_driver_ids = random.sample(driver_ids, num)
    df_of_some_drivers = pd.DataFrame(columns = df.columns)
    for driver_id in selected_driver_ids:
        data = df.loc[df['Driver_ID'] == driver_id]
        df_of_some_drivers = df_of_some_drivers.append(data)
    return df_of_some_drivers
```

```python
def convert_df_to_csv(df, file_name='new_data.csv'):
    df.to_csv(file_name)
    
```

```python
df_100_drivers = get_df_of_some_drivers(df,num=100)
```

```python
convert_df_to_csv(df_100_drivers, file_name='100_drivers_data.csv')
```
