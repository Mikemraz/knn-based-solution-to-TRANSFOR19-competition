```python
# some rows of data in prediction file are not given, 
# the file intends to created and interpolate thses rows
```

```python
import pandas as pd
from datetime import datetime
import numpy as np
```

```python
file_name = 'Predictions_south.csv'
prediction_north = pd.read_csv(file_name)
```

```python
pd.set_option('display.max_rows', 5000)
prediction_north
```

```python
def get_datetime_from_str(string):
    """get datetime of given string with format '%H:%M:%S'.
    
    Parameters:
    string: with format '%H:%M:%S'.
    
    Returns:
    datetime object
    
    """
    return datetime.strptime(string, '%H:%M:%S')

def get_str_from_datetime(date):
    """return string of datetime
    
    Parameters:
    date: datetime object
    
    Returns:
    string with format '%H:%M:%S'.
    
    """
    return date.strftime('%H:%M:%S')
```

```python
prediction_north.time = prediction_north.time.transform(get_datetime_from_str)
```

```python
prediction_north
```

```python
prediction_north_new = prediction_north.resample('5T', on='time').sum()
```

```python
prediction_north_new.speed = prediction_north_new.speed.replace(0, np.nan)
```

```python
prediction_north_new
```

```python
prediction_north_new.speed.loc['1900-01-01 00:00:00':'1900-01-01 05:55:00'] = \
prediction_north_new.speed.loc['1900-01-01 00:00:00':'1900-01-01 05:55:00'].interpolate('linear')
prediction_north_new.speed.loc['1900-01-01 11:00:00':'1900-01-01 15:55:00'] = \
prediction_north_new.speed.loc['1900-01-01 11:00:00':'1900-01-01 15:55:00'].interpolate('linear')
prediction_north_new.speed.loc['1900-01-01 21:00:00':'1900-01-01 23:55:00'] = \
prediction_north_new.speed.loc['1900-01-01 21:00:00':'1900-01-01 23:55:00'].interpolate('linear')
```

```python
prediction_north_new.index = [get_str_from_datetime(i) for i in prediction_north_new.index]
```

```python
prediction_north_new
```

```python
prediction_north_new.to_csv('cleaned_'+file_name)
```
