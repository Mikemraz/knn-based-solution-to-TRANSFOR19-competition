```python
import pandas as pd
import numpy as np
from datetime import datetime
from datetime import timedelta
from datetime import timezone
import matplotlib.pyplot as plt
from matplotlib.pyplot import xticks
from scipy import signal
```

```python
def add_datetime_column(row):
    """add a datetime column to a dataframe row.
    
    Parameters:
    row: a row from dataframe.
    
    Returns:
    date: datetime object
    """
    date = datetime.strptime(row['time'], '%Y-%m-%d %H:%M:%S')
    return date

def add_day_of_datetime_column(row):
    """add a column day of datetime to a dataframe row.
    
    Parameters:
    row: a row from dataframe
    
    Returns:
    day_of_datetime: the day of datetime, type int.
    
    """
    day_of_datetime = row['datetime'].day
    return day_of_datetime

def add_weekday_of_datetime_column(row):
    """add a column weekday of datetime to a dataframe row.
    
    Parameters:
    row: a row from dataframe
    
    Returns:
    weekday_of_datetime: weekday of datetime, 1 for Monday, 2 for Tuesday, and so on.
    
    """
    weekday_of_datetime = row['datetime'].weekday() + 1
    return weekday_of_datetime

def add_hour_min_second_column(row):
    """add column time of day to a dataframe row.
    
    Parameters:
    row: a row from dataframe
    
    Returns:
    hour_min_second: type string, with format '%H:%M:%S'.
    
    """
    hour_min_second = row['datetime'].strftime('%H:%M:%S')
    return hour_min_second

def add_hour_of_day_column(row):
    """add column hour of day to a dataframe row.
    
    Parameters:
    row: a row from dataframe.
    
    Returns:
    hour: hour of datetime.
    
    """
    hour = row['datetime'].strftime('%H')
    return hour

def add_order_num_of_point(row):
    """add column order_num of a datetime object.
    
    Parameters:
    row: a row from dataframe.
    
    """
    order_num = row['datetime'].day*3600*24 + row['datetime'].hour*3600 + row['datetime'].minute*60
    return order_num
```

```python
def replace_outliars(data,smooth_value=10):
    """replace the outliars in the speed data. 
    
    Parameters:
    smooth_value: the threshold value of smooth to replace outliars.
    
    Returns:
    data: the dataframe with outliars replaced.
    
    """
    column_index = allday_time_mean_speed.columns.get_loc('speed')
    for i in range(len(data)-2):
        index = i + 1
        value = data.iloc[index,column_index]
        previous_value = data.iloc[index-1,column_index]
        after_value = data.iloc[index+1,column_index]
        # the condition of detecting a outliar.
        if (abs(value-previous_value)>smooth_value) and (abs(value-after_value)>smooth_value):
            # recalculate the speed linear interpolation for outliars.
            data.iloc[index,column_index] = (previous_value+after_value)/2
    return data
```

```python
def preprocess(data):
    """ a wrap-up of all functions above.
    
    Parameters:
    data: the dataframe to be preprocessed
    
    Returns:
    data: preprocessed dataframe.
    """
    data = data.drop_duplicates('time')
    data = data.append({'time':'2016-10-01 00:00:04','speed':34,'number_of_drivers_considered':3},ignore_index=True)
    data['datetime'] = data.apply(lambda row:add_datetime_column(row),axis=1)
    data = data.resample('5T', on='datetime').sum()
    data['datetime'] = data.index
    data.loc[data.speed==0, 'speed'] = None
    data['speed'] = data['speed'].interpolate(mothods='time')
    data['day_of_datetime'] = data.apply(lambda row:add_day_of_datetime_column(row),axis=1,result_type='expand')
    data['weekday_of_datetime'] = data.apply(lambda row:add_weekday_of_datetime_column(row),axis=1)
    data['hour_min_second'] = data.apply(lambda row:add_hour_min_second_column(row),axis=1)
    data['order_num'] = data.apply(lambda row:add_order_num_of_point(row),axis=1)
    data['hour_of_day'] = data.apply(lambda row:add_hour_of_day_column(row),axis=1)
    data = data.sort_values('datetime')
    return data
```

```python
def visualize_certain_weekday_speed_trend(allday_data,weekday = 2):
    """plot the speed trend of a given weekday.
    
    Parameters:
    allday_data: the dataframe of all days' speed.
    weekday: the weekday to be plotted. 1 for Monday, 2 for Tuesday, and so on.
    
    Returns:
    None
    
    """
    weekday_data = allday_data[allday_data['weekday_of_datetime']==weekday]
    num_locs = 0
    for day_of_datetime in weekday_data['day_of_datetime'].unique():
        day_in_weekday_data = weekday_data[weekday_data['day_of_datetime']==day_of_datetime]
        lab = day_in_weekday_data['datetime'].iloc[0].strftime('%Y-%m-%d')
        plt.plot(day_in_weekday_data['hour_min_second'],day_in_weekday_data['speed'],label=lab)
    xticks(np.linspace(0, 288, num=13), ['{}:00'.format(i) for i in np.arange(0,25,step=2)],rotation=45)
    locs,labels = xticks()
    plt.legend(loc='upper right')
    plt.xlabel('time of day')
    plt.ylabel('speed(km/h)')
    plt.ylim(0,60)
    plt.title('day of weekday: {}'.format(weekday))
    plt.savefig('weekday_{}_speed.png'.format(weekday))
    plt.show()
    return None
```

```python
direction = 'north'
allday_time_mean_speed = pd.read_csv('300s_interval_to_to_{}_historical_target_road_average_speed_data.csv'.format(direction))
```

```python
allday_time_mean_speed = preprocess(allday_time_mean_speed)
```

```python
# output cleaned speed data.
output_file = allday_time_mean_speed[['speed']]
output_file.to_csv('to_{}_historical_road_speed_cleaned.csv'.format(direction))
```

```python
# plot extracted and cleaned speed data.
for weekday in range(1,8):
    visualize_certain_weekday_speed_trend(allday_time_mean_speed,weekday = weekday)
```
